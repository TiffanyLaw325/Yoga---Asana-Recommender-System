```python
#Import the libraries
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Load the data
data = pd.read_csv('/kaggle/input/yoga-asanas-and-its-benefits/final_asan1_1.csv')
data.head(5)
```

```python
#the summary of the data
data.info()
```

```python
#Create a list that wants to keep
columns = ['AName','Description','Benefits']
```

```python
#Create a function to combine the important columns/features
def combine_features(data):
    features =[]
    for i in range(0,data.shape[0]):
        features.append(str(data['AName'][i]) + ' '+str(data['Description'][i])+ ' '+str(data['Benefits'][i]))
        
    return features
```

```python
# Create a column to store the combine features
data['combine_features']= combine_features(data)
#Show the data
data.head(5)
```

```python
#Convert the text from combine_features to a matrix of word counts
cm =CountVectorizer().fit_transform(data['combine_features'])
```

```python
#Get the cosine similarity matrix from the word matrix
cs = cosine_similarity(cm)
#Print the scores
print(cs)
```

```python
#Get the Asana Name that Yogi likes
Asana = data['AName'][100]
#Show the title
Asana
```

```python
#Find the Asana ID that the Yogi likes
Asana_ID =data[data.AName==Asana]['AID'].values[0]
Asana_ID
```

```python
#Create a list of tuples in the form(AID, similarity)
scores = list(enumerate(cs[Asana_ID]))
```

```python
#Sort the list of similar Asanas in descending order
sorted_scores = sorted (scores, key = lambda x:x[1], reverse = True)
#Show the sorted scores
sorted_scores = sorted_scores[1:]
sorted_scores
```

```python
#Create a loop to print the first 5 Asanas from the sorted list
j = 0
print('The 5 most recommended asanas to '+Asana+' are:\n')
for item in sorted_scores:
    Asana_Name = data.loc[data['AID'] == item[0]]['AName']
    Level = data.loc[data['AID'] == item[0]]['Level']
    Asana_ben = data.loc[data['AID'] == item[0]]['Benefits']
    print(j+1,*Asana_Name,*Level,*Asana_ben,sep='\n')
    j=j+1
    if j >=6:
            break 
```
